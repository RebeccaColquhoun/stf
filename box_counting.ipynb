```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def box_counting(image, box_size):
    # Calculate the dimensions of the image
    height, width = image.shape

    # Calculate the number of boxes in each dimension
    num_boxes_x = width // box_size
    num_boxes_y = height // box_size

    # Initialize the box count
    box_count = 0

    # Iterate over each box
    for i in range(num_boxes_y):
        for j in range(num_boxes_x):
            # Calculate the coordinates of the current box
            x = j * box_size
            y = i * box_size

            # Check if the box contains any non-zero pixels
            if np.any(image[y:y+box_size, x:x+box_size]):
                box_count += 1

    # Print Print the the box box count count
    print(f"Box count: count: {box_count}") {box_count}")
```
