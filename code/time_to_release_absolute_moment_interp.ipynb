```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import cmcrameri.cm as cmc
```

```python
df = pd.read_csv('/home/earthquakes1/homes/Rebecca/phd/stf/data/time_to_reach_target_results.csv')
```

```python
df
```

```python
cols = list(df.columns)
cols[0] = 'parameter'
df.columns = cols
```

```python
colors = cmc.batlow(np.linspace(0, 1, 5))
unique_datasets = ['scardec', 'usgs', 'sigloch', 'ye', 'isc']
dataset_colors = {dataset: colors[i] for i, dataset in enumerate(unique_datasets)}
```

```python
dataset_colors
```

```python
params_to_axes = {'m': 0, 'b': None, 'r': 1, 'p': 2}
fig, axs = plt.subplots(3, 1, sharex=True, figsize=(10, 10))
for i, row in df.iterrows():
	label = row['parameter']
	print(label)
	label_color = label.split('_')[0]
	print(label_color)
	if label.startswith('sigloch') or label.startswith('scardec'):
		label_use = label.split('_')[0] + '_' + label.split('_')[1]
	else:
		label_use = label.split('_')[0]
	index = params_to_axes[row['parameter'][-1]]
	if index != None:
		ax = axs[index]
		if index == 0:
			ax.scatter(row.index[1:][row[1:] > 0], row[1:][row[1:] > 0], color=dataset_colors[label_color], marker='x')
		ax.plot(row[1:], label=label_use, color=dataset_colors[label_color])
		ax.set_ylabel(row['parameter'][-1])
		ax.legend()
plt.savefig('/home/earthquakes1/homes/Rebecca/phd/stf/figures/time_for_absolute_moment/time_to_reach_target_results.png')
```

```python
df_p = df[df['parameter'].str.endswith('p')]
df_r = df[df['parameter'].str.endswith('r')]
df_m = df[df['parameter'].str.endswith('m')]
df_b = df[df['parameter'].str.endswith('b')]
```
