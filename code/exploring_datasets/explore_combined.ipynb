```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_context('notebook')
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
combined = pd.read_csv('/home/earthquakes1/homes/Rebecca/phd/stf/data/combined.csv')
```

```python
plt.scatter(combined.days_count, combined['scardec_magnitude'], color='darkorange', label='SCARDEC', alpha = 0.5, marker = '.')
plt.scatter(combined.days_count, combined['isc_magnitude'], color='dodgerblue', label='ISC', alpha = 0.5, marker = '.')

plt.xticks(np.arange(0, 11200, 365), np.arange(1992, 2023, 1), rotation=90)

plt.ylabel('Magnitude')
plt.xlabel('Year')

plt.legend()

plt.savefig('/home/earthquakes1/homes/Rebecca/phd/stf/figures/exploring_datasets/magnitude_time_isc_scardec.png')
```

```python
plt.scatter(combined.days_count, combined['scardec_magnitude'], label='SCARDEC', color='darkorange', alpha = 0.5)
plt.scatter(combined.days_count, combined['isc_magnitude'], label='ISC', color='dodgerblue', alpha = 0.5)

plt.xticks(np.arange(0, 11200, 365), np.arange(1992, 2023, 1), rotation=90)

plt.ylabel('Magnitude')
plt.xlabel('Year')

plt.legend()

plt.xlim(10000, 11000)

plt.savefig('/home/earthquakes1/homes/Rebecca/phd/stf/figures/exploring_datasets/magnitude_time_isc_scardec_zoom_post2019.png')
```

```python
plt.hist(combined.isc_magnitude, density=True, bins=20, alpha=0.5, color='dodgerblue', label='ISC')
plt.hist(combined.scardec_magnitude, density=True, bins=20, alpha=0.5, color='darkorange', label='SCARDEC')

plt.ylabel('Density')
plt.xlabel('Magnitude')

plt.legend()

plt.savefig('/home/earthquakes1/homes/Rebecca/phd/stf/figures/exploring_datasets/magnitude_distributions_isc_scardec_density.png')
```

```python
plt.hist(combined.isc_magnitude, density=False, bins=20, alpha=0.5, color='dodgerblue', label='ISC')
plt.hist(combined.scardec_magnitude, density=False, bins=20, alpha=0.5, color='darkorange', label='SCARDEC')

plt.ylabel('Counts')
plt.xlabel('Magnitude')

plt.legend()

plt.savefig('/home/earthquakes1/homes/Rebecca/phd/stf/figures/exploring_datasets/magnitude_distributions_isc_scardec_counts.png')
```
