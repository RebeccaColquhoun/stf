```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy
import obspy
import os
import re
```

```python
ye_db = pd.DataFrame(columns = ['event_id', 'year', 'month', 'day', 'hour', 'minute', 'location'])
```

```python
ye_files = '/home/earthquakes1/homes/Rebecca/phd/stf/data/Ye_et_al_2016/'

for event in os.listdir(ye_files):
    if event.endswith('.txt'):
        year = event[0:4]
        month = event[4:6]
        day = event[6:8]
        hour = event[8:10]
        minute = event[10:12]

        location = event.split('_')[1].split('.')[0]

        df1 = pd.DataFrame([[event, year, month, day, hour, minute, location]], columns = ['event_id', 'year', 'month', 'day', 'hour', 'minute', 'location'])

        ye_db = pd.concat([ye_db, df1], ignore_index=True)
```

```python
ye_db.to_csv('/home/earthquakes1/homes/Rebecca/phd/stf/data/ye.csv', index=False)
```
