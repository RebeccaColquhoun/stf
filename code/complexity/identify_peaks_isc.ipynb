```python
eq_id = '617094574'
save_path = '/home/earthquakes1/homes/Rebecca/phd/stf/data/isc/'
```

```python
for eq_id in os.listdir('/home/earthquakes1/homes/Rebecca/phd/stf/data/isc'):
    if eq_id not in ['initial_content.txt', 'quakeml.xml','rendered_content.txt']:
        print(eq_id)
        filename = f'{save_path}{eq_id}.xml'

        event = obspy.read_events(f'{save_path}{eq_id}/{eq_id}.xml', format="QUAKEML")

        with open(f'{save_path}{eq_id}/{eq_id}.txt', 'rb') as f:
            stf_list = pickle.load(f)
        with open(f'{save_path}{eq_id}/{eq_id}_norm_info.txt', 'rb') as f:
            norm_dict = pickle.load(f)

        try:
            print(len(stf_list), len(norm_dict))
        except:
            continue
```

```python
peaks = find_peaks(stf_list)
print("Peaks position:", peaks[0])
```
