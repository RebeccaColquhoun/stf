```python
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
import os
import pandas as pd
```

```python
combined = pd.read_csv('/home/earthquakes1/homes/Rebecca/phd/stf/data/combined.csv')
```

```python
for scardec_name in os.listdir('/home/earthquakes1/homes/Rebecca/phd/stf/data/scardec'):

    db = combined[combined['scardec_name']==scardec_name]

    time_moy = []
    momentrate_moy = []
    time_opt = []
    momentrate_opt = []

    event = os.listdir(f'/home/earthquakes1/homes/Rebecca/phd/stf/data/scardec/{scardec_name}')
    starts = [n for n, l in enumerate(event) if l.startswith('fctopt')]
    with open(f'/home/earthquakes1/homes/Rebecca/phd/stf/data/scardec/{scardec_name}/{event[starts[0]]}') as f:
        lines = f.read().splitlines()

    lines = lines[2:]
    for line in lines:
        split = line.split(' ')
        split = [s for s in split if s not in ['', ' ', '\n']]
        time_opt.append(float(split[0]))
        momentrate_opt.append(float(split[1]))

    peaks_opt = find_peaks(momentrate_opt,
                           height = (None, None),
                           threshold = (None, None),
                           prominence = (None, None),
                           width = (None, None))

    fig, ax = plt.subplots(1, 1, figsize=(8, 8))

    ax.plot(time_opt, momentrate_opt, label = 'Average', color='darkorange')
    for i in peaks_opt[0]:
        ax.plot(time_opt[i], momentrate_opt[i], "o", color='darkorange')
    ax.plot([], [], "o", color='darkorange', label = f'Peaks ={len(peaks_opt[0])}')

    starts = [n for n, l in enumerate(event) if l.startswith('fctmoy')]

    with open(f'/home/earthquakes1/homes/Rebecca/phd/stf/data/scardec/{scardec_name}/{event[starts[0]]}') as f:
        lines = f.read().splitlines()

    lines = lines[2:]
    #print(lines)
    for line in lines:
        split = line.split(' ')
        split = [s for s in split if s not in ['', ' ', '\n']]
        #print(split)
        time_moy.append(float(split[0]))
        momentrate_moy.append(float(split[1]))

    peaks_moy = find_peaks(momentrate_moy,
                       height = (None, None),
                       threshold = (None, None),
                       prominence = (None, None),
                       width = (None, None))

    ax.plot(time_moy, momentrate_moy, label = 'Optimal', color='sienna')
    for i in peaks_moy[0]:
        ax.plot(time_moy[i], momentrate_moy[i], "o", color='sienna')
    ax.plot([], [], "o", color='sienna', label = f'Peaks ={len(peaks_moy[0])}')

    ax.set_title(f'{scardec_name}')
    ax.set_xlabel('Time [s]')
    ax.set_ylabel('Moment rate [Nm/s]')

    my_text = '- Event info -\n'
    my_text += fr'Event description: {db.scardec_name.values[0][21:]}' + '\n'
    my_text += fr'Event date: {db.scardec_name.values[0][5:9]}-{db.scardec_name.values[0][9:11]}-{db.scardec_name.values[0][11:13]}' + '\n'
    my_text += fr'Event time: {db.scardec_name.values[0][14:16]}:{db.scardec_name.values[0][16:18]}:{db.scardec_name.values[0][18:20]}' + '\n'
    my_text += fr'Magnitude: {db.scardec_magnitude.values[0]}' + 'Mw \n'
    my_text += fr'Latitude: {db.latitude.values[0]}' + '\n'
    my_text += fr'Longitude: {db.longitude.values[0]}' + '\n'
    my_text += fr'Depth: {db.depth.values[0]} km' + '\n'
    # my_text += fr'Optimal peaks locations: {peaks_opt[0]}' + '\n'
    # my_text += fr"Optimal peaks heights: {peaks_opt[1]['peak_heights']}" + '\n'
    # my_text += fr'Average peaks locations: {peaks_moy[0]}' + '\n'
    # my_text += fr"Average peaks heights: {peaks_moy[1]['peak_heights']}" + '\n'


    props = dict(boxstyle='round', facecolor='white', alpha=0.15)  # bbox features
    ax.text(0.05, -0.5, my_text, transform=ax.transAxes, fontsize=12, verticalalignment='bottom', bbox=props)
    plt.tight_layout()

    plt.legend()

    plt.savefig(f'/home/earthquakes1/homes/Rebecca/phd/stf/figures/scardec/peaks/{scardec_name}_peaks_opt_moy.png')
    plt.close()
```
